# Mapping Registration Numbers

```python

sc = Spark.sparkContext

reg_numbers = ["58001", "57002", "38003", "39004", "47005", "58006", "57007", "39008"]

rdd = sc.parallelize(reg_numbers)

def map_to_branch(reg_num):
    if reg_num.startswith('580'):
        return (reg_num, 'BDA')
    elif reg_num.startswith('570'):
        return (reg_num, 'AIML')
    elif reg_num.startswith('380'):
        return (reg_num, 'VLSI')
    elif reg_num.startswith('390'):
        return (reg_num, 'ES')
    elif reg_num.startswith('470'):
        return (reg_num, 'CDC')
    else:
        return (reg_num, 'Unknown')

branch_rdd = rdd.map(map_to_branch)

result = branch_rdd.collect()
for item in result:
    print(item)

```

```python

sc = Spark.sparkContext

rdd = sc.textFile("numbers.txt")

numbers_rdd = rdd.flatMap(lambda line: map(int, line.split()))

max_value = numbers_rdd.max()
min_value = numbers_rdd.min()
sum_value = numbers_rdd.sum()
count_value = numbers_rdd.count()
mean_value = sum_value / count_value if count_value != 0 else 0

print(f"Max: {max_value}")
print(f"Min: {min_value}")
print(f"Sum: {sum_value}")
print(f"Mean: {mean_value}")

```

# State Code Conversion

```python

citizen_rdd = sc.textFile("citizen.txt")

state_mapping = {
    "Karnataka": "KA",
    "Tamil Nadu": "TN",
    "Kerala": "KL",
}

def compress_state(line):
    parts = line.split(",")
    if len(parts) > 4:
        state_name = parts[4].strip()
        state_code = state_mapping.get(state_name, state_name)
        parts[4] = state_code 
    return ",".join(parts)

compressed_rdd = citizen_rdd.map(compress_state)

compressed_rdd.saveAsTextFile("citizen_compressed.txt")
```



# Student Data

```python

student_rdd = sc.textFile("student.txt")

data = student_rdd.map(lambda line: line.split(","))
```

```python
institute_count = data.map(lambda x: (x[1], 1)) \
                      .reduceByKey(lambda a, b: a + b)

print("Students from each Institute:")
for inst, count in institute_count.collect():
    print(inst, count)
```

```python
program_count = data.map(lambda x: (x[2], 1)) \
                    .reduceByKey(lambda a, b: a + b)

print("Students in each Program:")
for prog, count in program_count.collect():
    print(prog, count)
```

```python
gender_count = data.map(lambda x: (x[3], 1)) \
                   .reduceByKey(lambda a, b: a + b)

print("Gender Count:")
for gender, count in gender_count.collect():
    print(gender, count)
```

```python
selected_institute = "RVCE"

filtered_data = data.filter(lambda x: x[1] == selected_institute)

gender_selected = filtered_data.map(lambda x: (x[3], 1)) \
                                .reduceByKey(lambda a, b: a + b)

print(f"Gender count in {selected_institute}:")
for gender, count in gender_selected.collect():
    print(gender, count)
```

```python
temp_rdd = sc.textFile("temperature.txt")

data = temp_rdd.map(lambda line: line.split(","))
```

```python
temps = data.map(lambda x: float(x[1]))

max_temp = temps.max()
min_temp = temps.min()

print("Overall Maximum Temperature:", max_temp)
print("Overall Minimum Temperature:", min_temp)
```

```python
city_count = data.map(lambda x: (x[2], 1)) \
                 .reduceByKey(lambda a, b: a + b)

print("Data Points per City:")
for city, count in city_count.collect():
    print(city, count)
```

```python
bangalore_data = data.filter(lambda x: x[2].lower() == "bangalore")

bangalore_temps = bangalore_data.map(lambda x: float(x[1]))

print("Bangalore Maximum Temp:", bangalore_temps.max())
print("Bangalore Minimum Temp:", bangalore_temps.min())
```

```python
city_name = "Chennai"   # change city here

city_data = data.filter(lambda x: x[2].lower() == city_name.lower())

city_temps = city_data.map(lambda x: float(x[1]))

print(f"{city_name} Maximum Temp:", city_temps.max())
print(f"{city_name} Minimum Temp:", city_temps.min())
```

```python
bank_rdd = sc.textFile("Bank.txt")

data = bank_rdd.map(lambda line: line.split(","))
```

```python
unique_customers = data.map(lambda x: x[1]) \
                       .distinct() \
                       .count()

print("Unique Customers:", unique_customers)
```

```python
unique_banks = data.map(lambda x: x[0]) \
                   .distinct() \
                   .count()

print("Unique Bank IDs:", unique_banks)
```

```python
customers_per_bank = data.map(lambda x: (x[0], x[1])) \
                         .distinct() \
                         .map(lambda x: (x[0], 1)) \
                         .reduceByKey(lambda a, b: a + b)

print("Unique Customers per Bank:")
for bank, count in customers_per_bank.collect():
    print(bank, count)
```

```python
account_number = "1001"

transaction_count = data.filter(lambda x: x[1] == account_number) \
                        .count()

print(f"Total transactions for Account {account_number}:", transaction_count)
```

```python
account_number = "1001"
year = "2023"

credit_count = data.filter(lambda x: x[1] == account_number and
                                     x[3].lower() == "credit" and
                                     x[2].split("-")[2] == year) \
                   .count()

print(f"Credit transactions for Account {account_number} in {year}:", credit_count)
```
